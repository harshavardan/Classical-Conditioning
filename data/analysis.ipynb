```python
import pandas as pd
import matplotlib.pyplot as plt
filepath = "/home/harshavardan/Desktop/gitrepos/classical-conditioning/data/200921/F2/"
```

```python
trials = pd.read_csv(filepath + "trial.csv", header = None)[0]
trials
```

```python
precs = []
preus = []
train = []
probe = []
test = []

for i in range(len(trials)):
    if trials[i] == "precs":
        precs.append(i + 1)
    elif trials[i] == "preus":
        preus.append(i + 1)
    elif trials[i] == "training":
        train.append(i + 1)
    elif trials[i] == "probe":
        probe.append(i + 1)
    elif trials[i] == "test":
        test.append(i + 1)
```

```python
time = []
for i in range(800):
    time.append(i / 200.0 * 1000)
```

```python
keyword = "precs"
plt.figure(figsize = (20, 10))
for i in precs:
    plt.plot(time, pd.read_csv(filepath + "rawMotion" + str(i) + ".csv")[:800], color = "gray", alpha = 0.5)
    plt.ylim(0, 10)
    plt.xlabel("Time (ms)")
    plt.ylabel("Motion (arbitrary units)")
    plt.title(keyword)
    
avg = [0] * 800
for i in precs:
    k = 0
    for j in pd.read_csv(filepath + "rawMotion" + str(i) + ".csv")[pd.read_csv(filepath + "rawMotion" + str(i) + ".csv").columns[0]][:800]:
        avg[k] = avg[k] + j/len(precs)
        k = k + 1
        
plt.ylim(0, 5)
plt.xlim(0,4000)
plt.plot(time, avg[:800], "k")
plt.axvspan(0, 4000, alpha = 0.5, color = 'lightblue')

plt.savefig(filepath + keyword + ".png")
```
