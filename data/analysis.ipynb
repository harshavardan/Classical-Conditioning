```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
filepath = "./201009/F3/"
size = (20,10)
```

```python
trials = pd.read_csv(filepath + "trial.csv", header = None)[0]
```

```python
precs = []
preus = []
train = []
probe = []
test = []

for i in range(len(trials)):
    if trials[i] == "precs":
        precs.append(i + 1)
    elif trials[i] == "preus":
        preus.append(i + 1)
    elif trials[i] == "training":
        train.append(i + 1)
    elif trials[i] == "probe":
        probe.append(i + 1)
    elif trials[i] == "test":
        test.append(i + 1)
```

```python
time = []
for i in range(800):
    time.append(i / 200.0 * 1000)
```

```python
keyword = "precs"
plt.figure(figsize = size)
for i in precs:
    plt.plot(time, pd.read_csv(filepath + "rawMotion" + str(i) + ".csv")[:800], color = "gray", alpha = 0.5)
    plt.ylim(0, 10)
    plt.xlabel("Time (ms)")
    plt.ylabel("Motion (arbitrary units)")
    plt.title(keyword)
    
avg = [0] * 800
for i in precs:
    k = 0
    for j in pd.read_csv(filepath + "rawMotion" + str(i) + ".csv")[pd.read_csv(filepath + "rawMotion" + str(i) + ".csv").columns[0]][:800]:
        avg[k] = avg[k] + j/len(precs)
        k = k + 1
        
plt.ylim(0, 5)
plt.xlim(0,4000)
plt.plot(time, avg[:800], "k")
plt.axvspan(0, 4000, alpha = 0.5, color = 'lightblue')

plt.savefig(filepath + keyword + ".png")
```

```python
wait = pd.read_csv(filepath + "rawMotion0.csv", header = None)[0]
plt.plot(wait)
baseline = statistics.mean(wait[:50000])
sd = statistics.stdev(wait[:50000])
print(sd)
```

```python
keyword = "precs"
length = len(precs)
blah = [0, 0, 0, 0]
for i in precs:
    print(i)
    
    prob = [0, 0, 0, 0]
    
    time = pd.read_csv(filepath + "timestamp" + str(i) + ".csv", header = None)[0]
    b = time[0]
    for l in range(len(time)):
        time[l] = time[l] - b
        
    for j in range(800):
        
        if time[j] <= 1000:
            if pd.read_csv(filepath + "rawMotion" + str(i) + ".csv", header = None)[0][j] >= (baseline + 5 * sd):
                prob[0] += 1
                j = 200
        elif time[j] <= 2000:
            if pd.read_csv(filepath + "rawMotion" + str(i) + ".csv", header = None)[0][j] >= (baseline + 5 * sd):
                prob[1] += 1
                j = 400
        elif time[j] <= 3000:
            if pd.read_csv(filepath + "rawMotion" + str(i) + ".csv", header = None)[0][j] >= (baseline + 5 * sd):
                prob[2] += 1
                j = 600
        elif time[j] <= 4000:
            if pd.read_csv(filepath + "rawMotion" + str(i) + ".csv", header = None)[0][j] >= (baseline + 5 * sd):
                prob[3] += 1
                j = 800
    
    for k in range(4):
        if prob[k] > 0:
            blah[k] += 1/length
               
plt.figure(figsize = (20,10))
plt.bar(["0s-1s", "1s-2s", "2s-3s", "3s-4s"], blah, color = "k")
plt.ylabel("Probability of response")
plt.ylim(0,1.1)
plt.savefig(filepath + keyword + "probability" + ".png")
```
