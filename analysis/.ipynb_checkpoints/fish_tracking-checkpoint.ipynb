```python
import cv2
import numpy as np
import csv
import matplotlib.pyplot as plt
import statistics as stat
from scipy import ndimage
import tifffile as tif
```

# Background Estimation

```python
filename = "video2.avi"
location = "//storage.ncbs.res.in/harshavardanbn/data/210331/F3/"
length = 10000

count = 0
random = np.random.randint(300, length, 50)

bg_frames = [] #stores frames to used for bg calculation

cap = cv2.VideoCapture(location + filename)

if cap.isOpened == False:
    print("Unable to open file :/")

while True:
    
    ret,frame = cap.read()
    
    if ret == 0:
        break
    
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    if count == length:
        break
        
    if count in random:
        copy = frame.copy()
        bg_frames.append(copy)
    
    cv2.imshow("frame", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    count += 1
    
cap.release()
cv2.destroyAllWindows()

bg_frames = np.array(bg_frames)
bg = np.median(bg_frames, axis = 0)
cv2.imshow('Background', bg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Motion Detection

```python
cap = cv2.VideoCapture(location + filename)
count = 0
x_coor = []
y_coor = []
while cap.isOpened:
    no_contours = False    
    ret,frame = cap.read()
    
    if ret == 0:
        break
    if count == 10000:
         break
    
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    fg = frame - bg
    fg = cv2.convertScaleAbs(fg)
   # _, fg = cv2.threshold(fg, 100, 255, cv2.THRESH_BINARY)
    
    _, _, _, center = cv2.minMaxLoc(fg)
    cv2.circle(frame, center, 5, 255, 1)
    cv2.imshow("frame", ndimage.maximum_filter(fg, size = 5))
    x_coor.append(center[0])
    y_coor.append(center[1])
   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    count += 1
    
cap.release()
cv2.destroyAllWindows()
```

```python
plt.figure(figsize = (10, 10))
plt.scatter(x_coor, y_coor)

output = [x_coor, y_coor]
fields = ["x","y"]

with open(location + filename[:-4] + "coordinates.csv", "w") as csv_file:
    csvwriter = csv.writer(csv_file)
    csvwriter.writerow(fields)
    for i in range(len(output[0])):
        csvwriter.writerow([output[0][i], output[1][i]])
```

```python
dist = []
vel = []

for i in range(len(x_coor) - 1):
    distance = ((x_coor[i + 1] - x_coor[i])**2 + (y_coor[i + 1] - y_coor[i])**2)**0.5
    if distance <= 7:
        dist.append(0)
    else:
        dist.append(distance)
    #if i != 0:
        #dist[-1] += dist[-2]
```

```python
dist[0] = 0
plt.figure(figsize = (20, 10))
plt.plot(dist[:])
plt.xlabel("time (s)", fontsize = 15)
plt.ylabel("distance / frame (pixels)", fontsize = 15)
plt.savefig("./dist_time_plot.png")
```
