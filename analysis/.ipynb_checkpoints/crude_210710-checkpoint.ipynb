```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
import csv
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
 location = "D:/Harshavardan/data"
#location = "//storage.ncbs.res.in/harshavardanbn/data/tap_us/210711/F1"
files = 106
fps = 100
```

```python
all_trials = []
for i in range(1, files + 1):
    temp = []
    with open(location + "/rawmotion" + str(i) + ".csv", "r") as file:
        reader = csv.reader(file)
        for row in reader:
            temp.append(float(row[0]))
    all_trials.append(temp[:1200])
    plt.figure(figsize = (20, 0.5))
    plt.plot(all_trials[-1][:], "k")
    plt.xlim(0, 1200)
    #plt.ylim(0, 0.1)
    plt.title(i)
    plt.fill_between([500, 700], [5, 5], color = "lightblue")
    plt.fill_between([697.5, 700], [5, 5], color = "red")
    #plt.plot([1500, 1500], [0, 5], "r")
all_trials = np.array(all_trials)
```

```python
maximum = 0
for i in all_trials:
    if max(i) > maximum:
        maximum = max(i)

all_trials /= maximum
contrast = np.round((all_trials - np.min(all_trials)) / (np.max(all_trials) - np.min(all_trials)) * 255)
```

```python
plt.figure(figsize = (10, 10))
plt.imshow(contrast, aspect = "auto")#, cmap = cm.Blues)
#plt.savefig("heatmap.png")
plt.title(location)
#plt.plot([500, 500], [0, 111], "b")
```
